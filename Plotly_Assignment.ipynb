# 1. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.


```python
np.random.seed(20)
data1 = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df1 = pd.DataFrame(data1)

fig = px.density_heatmap(df1, x="Month", y="Day", z="Sales", nbinsx=5, nbinsy=30)

fig.update_layout(title="Variation in Sales across Different Months and Days")

fig.show()
```

# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivoted_df = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

fig = px.imshow(pivoted_df, color_continuous_scale=px.colors.sequential.Viridis)

fig.update_layout(
    title='Heatmap of Sales Across Different Months and Days',
    xaxis_title='Day',
    yaxis_title='Month'
)

fig.update_coloraxes(colorbar_title='Sales')

fig.show()
```

# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x^2 + y^2))

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='viridis')])

fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'))

fig.show()
```
