```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

```python
x = np.random.uniform(0, 10, 100)
y = 2 * x + np.random.normal(0, 2, 100)

df = pd.DataFrame({'x': x, 'y': y})

sns.scatterplot(data=df, x="x", y="y", color = 'red' )

plt.title("Scatter Plot of X vs Y")
plt.show()
```

# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
numbers = np.random.rand(1000)

sns.histplot(numbers, bins=20, color = 'g')

plt.title("Distribution of Random Numbers")
plt.show()
```

# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

```python
data = {'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
        'Value': [10, 15, 12, 11, 16, 13]}
df = pd.DataFrame(data)

sns.barplot(x='Category', y='Value', data=df)

plt.title("Comparison of Categories Based on Numerical Values")
plt.show()
```

# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
categories = ['A', 'B', 'C']
data = {'Category': np.random.choice(categories, 100, replace=True),
        'Value': np.random.normal(0, 1, 100)}
df = pd.DataFrame(data)

sns.boxplot(x='Category', y='Value', data=df)

plt.title("Distribution of Numerical Variable Across Categories")
plt.show()
```

# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
data = pd.DataFrame({
    'Variable1': np.random.randn(100),
    'Variable2': np.random.randn(100) * 1.5 + 0.5,
    'Variable3': np.random.randn(100) * 0.5 - 0.5,
    'Variable4': np.random.randn(100) * 0.2 - 0.1
})

correlation_matrix = data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title('Correlation Matrix of Variables')
plt.show()
```
