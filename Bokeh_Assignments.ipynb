(Use Bokeh for the visualization of the given questions)

```python
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook, show
from bokeh.palettes import Spectral5, Spectral6
from bokeh.transform import factor_cmap, linear_cmap
from bokeh.embed import components
from bokeh.models import HoverTool
```

# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

source = ColumnDataSource(data=dict(x=x, y=y))

p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")

p.line(x='x', y='y', source=source, line_width=2)

output_notebook()
show(p)
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
np.random.seed(25)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(5, 20, 100)
colors = np.random.choice(['red', 'green', 'blue', 'yellow'], 100)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Scatter Plot with Different Marker Sizes and Colors")
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

output_notebook()

show(p)
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

data = {'Fruit': fruits, 'Count': counts}
df = pd.DataFrame(data)

p = figure(x_range=fruits, height=350, toolbar_location=None, title="Fruit Counts")
p.vbar(x='Fruit', top='Count', width=0.9, source=df, line_color='white', 
       fill_color=factor_cmap('Fruit', palette=Spectral5, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 40

output_notebook()

show(p)
```

# 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data", tools="", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="#5555ee", line_color="white", alpha=0.5)

hover = HoverTool(tooltips=[
    ("index", "$x"),
    ("value", "$y"),
])
p.add_tools(hover)

p.yaxis.axis_label = 'Frequency'
p.xaxis.axis_label = 'Value'

output_notebook()

show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.

```python
data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

df = pd.DataFrame({'x': xx.flatten(), 'y': yy.flatten(), 'z': data_heatmap.flatten()})

source = ColumnDataSource(df)

mapper = linear_cmap(field_name='z', palette=Spectral6, low=min(data_heatmap.flatten()), high=max(data_heatmap.flatten()))

p = figure(x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

p.rect(x='x', y='y', width=0.1, height=0.1, fill_color=mapper, line_color=None, source=source)

output_notebook()

show(p)
```
